# Data Visualization and Web Scraping

Seamlessly integrate data visualization and web scraping capabilities into your Python projects. This setup includes essential libraries for creating insightful visualizations and extracting data from web pages.

```python
# Quick import test and demo plot
import seaborn as sns
```

```python
### Basic Plotting with Seaborn
tips = sns.load_dataset("tips")
tips
```

```python
## Crete a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", sizes=(20, 200))
plt.title("Tips Scatter Plot")
plt.show()
```

```python
## Line plot
sns.lineplot(data=tips, x="total_bill", y="tip", hue="day", style="time", markers=True, dashes=False)
plt.title("Tips Line Plot")
plt.show()
```

```python
## Categorical plot
## Bar Plot
sns.catplot(data=tips, x="day", y="total_bill", kind="bar")
plt.title("Total Bill by Day")
plt.show()
```

```python
## Box Plot
sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="smoker")
plt.title("Total Bill by Day and Smoking Status")
plt.show()
```

```python
## Violin Plot
sns.catplot(data=tips, x="day", y="total_bill", kind="violin", hue="smoker", split=True)
plt.title("Total Bill by Day and Smoking Status - Violin Plot")
plt.show()
```

```python
### Histogram
sns.histplot(data=tips, x="total_bill", hue="smoker", multiple="stack")
plt.title("Total Bill Distribution by Smoking Status")
plt.show()
```

```python
## Pairplot
sns.pairplot(tips, hue="smoker", diag_kind="kde")
```

```python
## HEatmap
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
