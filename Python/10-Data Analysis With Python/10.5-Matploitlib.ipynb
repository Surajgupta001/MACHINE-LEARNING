# Data Visualization with Matplotlib

Matplotlib is a powerful library in Python for creating static, animated, and interactive visualizations. Below are some common types of plots you can create using Matplotlib.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# create a customized line plot
plt.plot(x, y, marker='o', linestyle='--', color='r')
plt.grid(True)
```

```python
## Multiple Plots
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9, 5))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, y1, color='green')
plt.title('plot 1')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.plot(y1, x, color='blue')
plt.title('plot 2')
```

```python
## Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## Histogram

Histogram are used to represent the distribution of dataset. They divided the data into bins and count th number of data points in each bin

```python
## Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='purple', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
## Create a Scatter Plot
# sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plt.scatter(x, y, color='orange', marker='x')
plt.title('Scatter Plot Example')
```

```python
## Pie chat
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # explode the 1st slice -> move it out
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
```
