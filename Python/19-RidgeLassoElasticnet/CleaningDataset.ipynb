# Algerian Forest Fires Dataset

Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.


122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 outp attribue (class) The 244 instances have been classified into fire (138 classes) and not fire (106) classes) classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning

```python
## Missing data
dataset[dataset.isnull().any(axis=1)]
```

The Dataset is converted into two sets based on region from 122th index, we can make a new based on the region

1 : Bejaia Region Dataset

2 : Sidi Bel-abbes Region Dataset

Add new column with regions

```python
dataset.loc[:122, 'Region'] = 0
dataset.loc[122:, 'Region'] = 1
df=dataset
```

```python
df.info()
```

```python
# Now convert to int
df[['Region']] = df[['Region']].astype('int')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

### Changes the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.head()
```

```python
### Changing the others columns to float data datatype
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype('float')
```

```python
df.info()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

## Exploratory Data Analysis

```python
# Dropping the day, month, year columns not needed for analysis
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
## Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features
plt.style.use('seaborn-v0_8-darkgrid')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
## percentage for of pie chart
percentabge = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting pie chart
classlabels = ['Not Fire', 'Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentabge, labels=classlabels, autopct='%1.1f%%', startangle=20, colors=['lightblue', 'orange'], explode=(0.05, 0.05))
plt.title('Percentage of Fire vs Not Fire Cases', fontsize=16)
plt.show()
```

```python
## Correlation
df_copy.corr()
```

```python
## Encoding of the categories in classes
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
## Monthly fire analysis
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Month', weight='bold')
plt.title('Monthly Fire Analysis in Sidi-Bel Region', weight='bold', size=16)
```

```python
## Monthly fire analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Month', weight='bold')
plt.title('Monthly Fire Analysis in Bajaie Region', weight='bold', size=16)
```
