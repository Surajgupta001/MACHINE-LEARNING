# SQL AND SQLite

SQL(Structure Query Language) is a standard language used to communicate with and manipulate databases. SQLite is a lightweight, disk-based database that doesn't require a separate server process and allows access to the database using a nonstandard variant of the SQL query language.

```python
import sqlite3
```

```python
## connect to an sqlite database
connection = sqlite3.connect('my_database.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS  employees(
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER,
        department TEXT
    )                      
''')

## Commit the changes
connection.commit()
```

```python
## Insert the data in the sqllite table
cursor.execute('''
    INSERT INTO employees (name, age, department)
    VALUES ('John Doe', 30, 'HR')
''')

## Commit the changes
connection.commit()
```

```python
## Query the data
cursor.execute('SELECT * FROM employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
## Upadte the data
cursor.execute('''
    UPDATE employees
    SET age = 31
    WHERE name = 'John Doe'
''')

## Commit the changes
connection.commit()
```

```python
cursor.execute('SELECT * FROM employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
## Delete the data
cursor.execute('''
    DELETE FROM employees
    WHERE name = 'John Doe'
''')

## Commit the changes
connection.commit()
```

```python
cursor.execute('SELECT * FROM employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
## Create a table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS  employees(
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER,
        department TEXT
    )                      
''')

## Commit the changes
connection.commit()
```

```python
cursor.execute(
    '''
    INSERT INTO employees (name, age, department)
    VALUES ('John Doe', 30, 'HR')
    '''
)

cursor.execute(
    '''
    INSERT INTO employees (name, age, department)
    VALUES ('Jane Smith', 25, 'IT')
    '''
)

cursor.execute(
    '''
    INSERT INTO employees (name, age, department)
    VALUES ('Mike Johnson', 40, 'Finance')
    '''
)

## Commit the changes
connection.commit()
```

```python
cursor.execute('SELECT * FROM employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
## close the connection
connection.close()
```

```python
## Working with sales data
## Connect to an SQLite database
connection = sqlite3.connect('sales_database.db')
cursor = connection.cursor()

## Create a sales table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales(
        id INTEGER PRIMARY KEY,
        product_name TEXT NOT NULL,
        quantity INTEGER,
        price REAL,
        sale_date TEXT
    )                      
''')

## Commit the changes
connection.commit()
```

```python
## Insert data into the table
cursor.execute('''
    INSERT INTO sales (product_name, quantity, price, sale_date)
    VALUES ('Laptop', 2, 1200.50, '2024-01-15')
''')

cursor.execute(
    '''
    INSERT INTO sales (product_name, quantity, price, sale_date)
    VALUES ('Smartphone', 5, 800.00, '2024-01-16')
    '''
)

cursor.execute(
    '''
    INSERT INTO sales (product_name, quantity, price, sale_date)
    VALUES ('Tablet', 3, 450.75, '2024-01-17')
    '''
)

cursor.execute(
    '''
    INSERT INTO sales (product_name, quantity, price, sale_date)
    VALUES ('Headphones', 10, 150.25, '2024-01-18')
    '''
)

connection.commit()
```

```python
cursor.execute('SELECT * FROM sales')
rows = cursor.fetchall()
for row in rows:
    print(row)
```
