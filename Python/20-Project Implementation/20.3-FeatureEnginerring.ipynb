```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## Missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, 'Region'] = 0
dataset.loc[122:, 'Region'] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype('int64')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Changing the required columns to integer datatypes

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype('int64')
```

```python
df.info()
```

## Changes the required columns to float datatypes

```python
df.head()
```

```python
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
objects
```

```python
for i in df.columns:
    if i not in ['Classes', 'Region']:
        df[i] = df[i].astype('float64')
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_dataset.csv', index=False)
```

## Exploratory Data Analysis

```python
df_copy = df.drop(['day', 'month', 'year'], axis = 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features
plt.style.use('seaborn-v0_8-darkgrid')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
## percentage for of pie chart
percentabge = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting pie chart
classlabels = ['Not Fire', 'Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentabge, labels=classlabels, autopct='%1.1f%%', startangle=20, colors=['lightblue', 'orange'], explode=(0.05, 0.05))
plt.title('Percentage of Fire vs Not Fire Cases', fontsize=16)
plt.show()
```
