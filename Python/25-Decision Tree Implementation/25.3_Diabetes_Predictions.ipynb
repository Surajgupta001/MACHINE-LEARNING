```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset.DESCR)
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data, columns=[
    'age',
    'sex',
    'bmi',
    'bp',
    's1',
    's2',
    's3',
    's4',
    's5',
    's6'
])
df_diabetes.head()
```

```python
# Independent and dependent features
X = df_diabetes
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
# Correlation
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyper-parameter Tuning

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [2, 3, 4, 5, 6, 7, 8, 9, 10],
    # 'auto' is not valid for DecisionTreeRegressor
    'max_features': [None, 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(estimator=regressor, param_grid=param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print('R2 Score:', r2_score(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
```

```python
regressor = DecisionTreeRegressor(criterion='friedman_mse', max_depth=4, max_features='log2', splitter='random')
```

```python
regressor.fit(X_train, y_train)
```

```python
# Visualization of the tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(regressor, filled=True)
```
