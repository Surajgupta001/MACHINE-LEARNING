## Principal Component Analysis (PCA) :
- PCA is a dimensionality reduction technique that transforms a large set of variables into a smaller one that still contains most of the information in the large set. It does this by identifying the directions (principal components) along which the variance of the data is maximized. The first principal component captures the most variance, followed by the second, and so on. PCA is commonly used for data visualization, noise reduction, and feature extraction in machine learning.
- PCA works by computing the covariance matrix of the data, finding its eigenvalues and eigenvectors, and then projecting the data onto the eigenvectors corresponding to the largest eigenvalues. This results in a new set of uncorrelated variables (principal components) that can be used for further analysis or modeling.
- PCA is sensitive to the scale of the data, so it is often recommended to standardize the data before applying PCA. This can be done by centering the data (subtracting the mean) and scaling it (dividing by the standard deviation) to ensure that each variable contributes equally to the analysis.
- PCA can be used for various applications, such as image compression, gene expression analysis, and market research. It helps in reducing the dimensionality of the data while retaining as much information as possible, making it easier to visualize and analyze complex datasets.
- formula for PCA:
1. Standardize the data:  
    \[
    X_{\text{standardized}} = \frac{X - \mu}{\sigma}
    \]
    where \(X\) is the original data, \(\mu\) is the mean, and \(\sigma\) is the standard deviation.

2. Compute the covariance matrix:  
    \[
    C = \frac{1}{n-1} X_{\text{standardized}}^T X_{\text{standardized}}
    \]
    where \(n\) is the number of samples.

3. Compute the eigenvalues and eigenvectors of the covariance matrix:  
    \[
    C v = \lambda v
    \]
    where \(v\) is the eigenvector and \(\lambda\) is the corresponding eigenvalue.

4. Sort the eigenvalues and their corresponding eigenvectors in descending order.

5. Select the top \(k\) eigenvectors to form the projection matrix \(W\):  
    \[
    W = [v_1, v_2, \ldots, v_k]
    \]

6. Project the standardized data onto the new subspace:  
    \[
    X_{\text{PCA}} = X_{\text{standardized}} W
    \]
    \]
- PCA is a powerful tool for reducing the dimensionality of data while preserving as much variance as possible. However, it is important to note that PCA assumes linear relationships between variables and may not capture complex nonlinear patterns in the data. Additionally, PCA can be sensitive to outliers, which can significantly affect the results. Therefore, it is crucial to preprocess the data appropriately and consider alternative techniques if the assumptions of PCA are not met.
- overall PCA is a widely used technique in data analysis and machine learning for dimensionality reduction, feature extraction, and data visualization. It helps to simplify complex datasets while retaining the most important information, making it easier to analyze and interpret the data. However, it is essential to understand the assumptions and limitations of PCA to ensure that it is applied correctly and effectively in various applications. 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
## Load the datasets
from sklearn.datasets import load_breast_cancer
```

```python
cancer_datasets = load_breast_cancer()
```

```python
cancer_datasets.keys()
```

```python
print(cancer_datasets.DESCR)
```

```python
df=pd.DataFrame(
    cancer_datasets.data, columns=cancer_datasets.feature_names
)
```

```python
df.head()
```

```python
# 1. Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
# Applying PCA Algorithm
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) ## We want to reduce the dimensions to 2
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=data_pca[:, 0], y=data_pca[:, 1], hue=cancer_datasets.target, palette='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Breast Cancer Dataset')
```
