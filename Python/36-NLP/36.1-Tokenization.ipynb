```python
corpus = """
Hello, Welcome to the world of Machine Learning!
NLTK is a powerful library for natural language processing.
It provides easy-to-use interfaces for tokenization, stemming, and more.
With NLTK, you can analyze text data and build applications like chatbots, sentiment analysis, and language translation.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)
print(sentences)
```

```python
type(sentences)
```

```python
for sentence in sentences:
    print(sentence)
```

```python
## Tokenization
# paragraphs ---> sentences
# sentences ---> words
from nltk.tokenize import word_tokenize

words = word_tokenize(corpus)
print(words)
```

```python
for word in words:
    print(word)
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
