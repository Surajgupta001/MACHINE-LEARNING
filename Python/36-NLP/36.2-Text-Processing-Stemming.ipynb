# Stemming
- Stemming is the process of reducing a word to its stem that suffixes and prefixes or the roots of words knowns as a lemma.
- Stemming is important in natural language understanding (NLU) and natural language processing (NLP) because it helps to reduce the number of unique words in a text, which can improve the performance of machine learning models and make it easier to analyze and understand the text.

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews ----> eating, eat, eaten, eats => eat is root word
## Stemming is the process of reducing a word to its root word

words = [
    "eating",
    "eat",
    "eaten",
    'writing',
    'writes',
    'programming',
    'programs',
    'history',
    'finally',
    'finalized'
]
```

## Porter Stemmer
- The Porter Stemmer is a widely used stemming algorithm that was developed by Martin Porter in 1980. It is a rule-based algorithm that uses a series of rules to reduce words to their stems.
- The Porter Stemmer is designed to be simple and efficient, and it is often used in information retrieval and natural language processing applications. It is based on a set of rules that are applied to the end of words to remove common suffixes and prefixes, such as "ing", "ed", "s", and "es". The algorithm also includes rules for handling irregular verbs and other exceptions.
- The Porter Stemmer is a widely used stemming algorithm that is implemented in many natural language processing libraries, including NLTK in Python. It is a popular choice for stemming because it is simple and efficient, and it can be used to reduce the number of unique words in a text, which can improve the performance of machine learning models and make it easier to analyze and understand the text.

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

for word in words:
    print(word, "---->", ps.stem(word))
```

```python
ps.stem('congratulation')
```

```python
ps.stem('sitting')
```

## RegexpStemmer class
- NLTK has RegexpStemmer class with the help of which we can create our own stemming rules. It takes a regular expression as an argument and applies it to the end of words to remove suffixes and prefixes. This allows us to create custom stemming rules that are specific to our needs.

```python
from nltk.stem import RegexpStemmer

rs = RegexpStemmer('ing$|s$|ed$|ables$', min=4)
```

```python
rs.stem('eating')
```

```python
rs.stem('ingeating')
```

## Snowball Stemmer
- The Snowball Stemmer is an improvement over the Porter Stemmer that was developed by Martin Porter in 2001. It is a more aggressive stemming algorithm that uses a larger set of rules to reduce words to their stems.
- The Snowball Stemmer is designed to be more aggressive than the Porter Stemmer, and it is often used in information retrieval and natural language processing applications. It is based on a larger set of rules that are applied to the end of words to remove common suffixes and prefixes, such as "ing", "ed", "s", and "es". The algorithm also includes rules for handling irregular verbs and other exceptions.
- The Snowball Stemmer is a widely used stemming algorithm that is implemented in many natural language processing libraries, including NLTK in Python. It is a popular choice for stemming because it is more aggressive than the Porter Stemmer, and it can be used to reduce the number of unique words in a text, which can improve the performance of machine learning models and make it easier to analyze and understand the text.

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer('english')
```

```python
for word in words:
    print(word, "---->", ss.stem(word))
```

```python
ps.stem('fairly'), ps.stem('sportingly')
```

```python
ss.stem('fairly'), ss.stem('sportingly')
```

```python
ss.stem('going')
```

```python
ss.stem('goes')
```
