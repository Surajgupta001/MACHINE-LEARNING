# Wordnet Lemmatizer
- Lemmatization technique is like stemming. The output we will get after lemmatization is called `lemma`, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides a WordNetLemmatizer class which is a thin wrapper around the WordNet corpus. This class uses `morphy()` function to find the lemma of a word. The `morphy()` function takes a word as input and returns its lemma if it is found in the WordNet corpus, otherwise it returns the original word.

```python
import nltk
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
```

```python
'''
POS- Noun-n
Verb-v
Adjective-a
Adverb-r
'''
wnl.lemmatize('going', pos='v')
```

```python
words = [
    "eating",
    "eat",
    "eaten",
    'writing',
    'writes',
    'programming',
    'programs',
    'history',
    'finally',
    'finalized'
]
```

```python
for word in words:
    print(word, '---->', wnl.lemmatize(word, pos='v'))
```

```python
wnl.lemmatize('goes', pos='v')
```
