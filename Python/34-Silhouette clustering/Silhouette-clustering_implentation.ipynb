## Silhouette (clustering)
- The silhouette is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). It ranges from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. A silhouette close to 0 indicates that the object is on or very close to the decision boundary between two neighboring clusters, while a negative silhouette indicates that the object might have been assigned to the wrong cluster.

```python
import pandas as pd
df = pd.read_csv('healthcare.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Healthcare Data')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(contamination=0.1)
clf.fit(df)
predictions = clf.predict(df)
```

```python
predictions
```

```python
import numpy as np
index = np.where(predictions < 0)
index
```

```python
x = df.values
```

```python
index = np.where(predictions < 0)
plt.scatter(x[:, 0], x[:, 1], color='blue', label='Normal')
plt.scatter(x[index, 0], x[index, 1], color='red', label='Anomalies')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomaly Detection using Isolation Forest')
plt.legend()
plt.show()
```
