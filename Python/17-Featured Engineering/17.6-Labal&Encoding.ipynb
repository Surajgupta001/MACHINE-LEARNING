# Label Encoding

Label encoding and ordinal are two techniques used to encode categorical variables into numerical format for machine learning algorithms.

Label ecoding involves assigning a unique numerical label to each category in a categorical variable. The labels are usually assigned in an alphabetical order or based on the frequency of occurrence of each category. For example, if we have a categorical variable "Color" with three categories: "Red", "Blue", and "Green", we can assign the following labels:

1. Red -> 0
2. Blue -> 1
3. Green -> 2

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df['Color'])
```

```python
lbl_encoder.transform(['Red'])
```

# Ordinal Encoding

It is used to encode categorical data that have an instrinsic order or ranking among the categories. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have a categorical variable "education level" with four possible values ("High School", College", "Graduate", "Postgraduate") we can assign the following ordinal values:

1. High School -> 1
2. College -> 2
3. Graduate -> 3
4. Postgraduate -> 4

```python
### ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'Education': ['High School', 'College', 'Graduate', 'Postgraduate', 'College']
})
```

```python
## Create an instance of OrdinalEncoder and then fit_transform the
encoder = OrdinalEncoder(categories=[['High School', 'College', 'Graduate', 'Postgraduate']])
```

```python
encoder.fit_transform(df[['Education']])
```

```python
encoder.transform([['Graduate']])
```
