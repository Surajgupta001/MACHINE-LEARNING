# SMOTE(Synthetic Minority Over-sampling Technique)

SMOTE is an advanced technique used in machine leaning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instance of the minority class by interpolating between existing minority class instances. This helps to create a more balanced dataset, which can improve the performance of machine learning models.

```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_clusters_per_class=1,
    weights=[0.90],
    random_state=12,
    n_redundant=0,
)
```

```python
df1 = pd.DataFrame(X, columns=["f1", "f2"])
df2 = pd.DataFrame(y, columns=["target"])

final_df = pd.concat([df1, df2], axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset
oversample = SMOTE()
x,y = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```
