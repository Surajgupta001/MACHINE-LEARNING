# Handling Imbalanced Dataset

1. Up sampling
2. Down sampling

```python
import numpy as np
import pandas as pd

#set the random seed for reproductibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9 # 90% of samples belong to class 0
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
## Create my dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=5, scale=2, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=3, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=8, scale=2, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
## Upsampling
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority, replace=True, # sample with replacement
                                 n_samples=len(df_majority), 
                                 random_state=42
                                 )    
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_unsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_unsampled['target'].value_counts()
```

## Down sampling

```python
import numpy as np
import pandas as pd

#set the random seed for reproductibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9 # 90% of samples belong to class 0
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

## Create my dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=5, scale=2, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=3, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=8, scale=2, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
df['target'].value_counts()
```

```python
## Down-sampling
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample

df_majority_downsampled = resample(df_majority, replace=False, # sample without replacement
                                 n_samples=len(df_minority), 
                                 random_state=42
                                 )    
```

```python
df_majority_downsampled.shape
```

```python
df_downSampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downSampled['target'].value_counts()
```
