## Reference counting
Reference counting is the primary method Python uses to manage memory. Each object in Python maintains a count of references pointing to it. When the reference count drops to zero, the memory occupied by the object is immediately deallocated.

```python
### Reference counting
import sys
a = []
print(sys.getrefcount(a))  # Output: 2 (one for 'a' and one for getrefcount argument)
```

```python
b = a
print(sys.getrefcount(b)) # Output: 3 (now 'a' and 'b' both reference the same list)
```

```python
del b
print(sys.getrefcount(a))  # Output: 2 (back to original reference count)
```

### Garbage Collection
Python includeds a cyclic garbage collector to handle reference cycles. Reference cycles occur when objects reference each other, preventing their reference counts from ever reaching zero. The garbage collector periodically identifies and collects these cycles to free up memory.

```python
import gc
## enable automatic garbage collection
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect()  # Manually trigger garbage collection
```

```python
### Get garbage Collection stats
print(gc.get_stats())
```

```python
### get unreachable objects
print(gc.garbage)
```

## Memory Management Best Practices

1. Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables.
2. Avoid Circular References: Circular references can lead to memory leaks if not handled properly.
3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making tehm memory efficient.
4. Explicitly Delete Objects: Use the del statement to delete variables and objects explicitly.
5. Profile Memory Usage: Use memory profiling tools like tracemalloc to memory_profiler to identify memory leaks and optimize memory usage. 

```python
# Handling Circular References
import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created.")
        
    def __del__(self):
        print(f"Object {self.name} deleted.")

# Create circular reference
obj1 = MyObject("obj1")
obj2 = MyObject("obj2")
obj1.ref = obj2
obj2.ref = obj1

# Remove references
del obj1
del obj2

# Manually trigger garbage collection
gc.collect()
```

## Generators for memory efficiency

Generators allow you to produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

```python
def generator_numbers(n):
    for i in range(n):
        yield i

## Using the generator
for num in generator_numbers(10000):
    print(num)
    if num > 10:
        break
```

```python
### Profile Memory Usage with tracemalloc
import tracemalloc

def create_list():
    return [i for i in range(100000)]  # Create a large list

def main():
    tracemalloc.start()  # Start tracing memory allocations
    
    create_list()  # Call the function that uses memory
    
    snapshot = tracemalloc.take_snapshot()  # Take a snapshot of memory usage
    top_stats = snapshot.statistics('lineno')  # Get statistics by line number
    
    print("[ Top 10 memory consuming lines ]")
    for stat in top_stats[::]:
        print(stat)
```

```python
main()
```
